```python
from ipynb.fs.full.ScPipelineProto import *
import cfad
from pathlib import Path
import math
```

```python
def split_channels_by_region(df):
    channels = {}
    unique_regions = df["acronym"].unique()

    for region in unique_regions:
        grouped = df.groupby("acronym")
        region_df = grouped.get_group(region)

        channel_ids = region_df["channel_id"].values

        channels.update({region: channel_ids})
    return channels
        
```

```python
results = []
Group = [None, "Vehicle", "Vehicle", "Psilocybin", "Vehicle", "Psilocybin", "Vehicle", "Psilocybin", "Psilocybin", "Vehicle"]
for mouse in [2, 3, 4, 5, 7]:
    trial_types = ["All", "Flash", "Loom", "EveryLoom"]
    experiments = ["extinction", "renewal"]
    for trial_type in trial_types:
        for experiment in experiments:
            if experiment == "extinction":
                mean_path = rf"Z:\Saij\Data\extinction\Mouse{mouse}\data\meanLFP{trial_type}\meanLFP.dat"
                experiment = "Extinction"
            else:
                mean_path = rf"Z:\Saij\Data\renewal\Mouse{mouse}\data\meanLFP{trial_type}\meanLFP.dat"
                experiment = "Retention"
            ccf_path = rf"Z:\Saij\ccf_channels\Mouse {mouse}.csv"
            df = pd.read_csv(ccf_path)
            
            region_channels = split_channels_by_region(df)
            
            mean_clean, mean_meta = load_scaled_lfp_from_dat(mean_path)
            
            fs = mean_meta
            if trial_type == "EveryLoom":
                window = (-0.1, 0.7)
                stims = [0]
                post = 0.5
            else:
                window = (-0.1, 5)
                post = 1
                if trial_type == "All":
                    stims = [0, 2.94]
                elif trial_type == "Loom":
                    stims = [0, 1.47, 2.94]
            offset = np.abs(window[0])
            stims_offset = stims + offset
            stims_samples = stims_offset * fs
            
            
        
            for region in region_channels:
                channels = region_channels[region]
                roi = mean_clean[channels]
                for n, stim in enumerate(stims_samples):
                    if trial_type == "Loom":
                        stim_type = f"Loom {n + 1}"
                    elif trial_type == "All":
                        if n == 0:
                            stim_type = "Flash On"
                        else:
                            stim_type = "Flash Off"
                    else:
                        stim_type = "Loom"
                    start = int(stim)
                    end = int(stim + (1 * mean_meta))
                    subset = roi[:, start:end]
                    
                    ind_min = np.unravel_index(np.nanargmin(subset), subset.shape)
                    min_value = subset[ind_min]
                    channel_min, sample_min = ind_min
                    latency_min = sample_min/1000
            
                    
                    ind_max = np.unravel_index(np.nanargmax(subset), subset.shape)
                    max_value = subset[ind_max]
                    channel_max, sample_max = ind_max
                    latency_max = sample_max/1000
            
                    values = [mouse, Group[mouse], experiment, region, stim_type, channel_min, min_value, latency_min, channel_max, max_value, latency_max]
                    results.append(values)

        

results_df = pd.DataFrame(results, columns = ["Mouse", "Treatment Group", "Experiment", "Region", "Stim", "Channel min", "min", "latency min", "Channel max", "max", "Latency max"])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# 1. Setup Context and Filter Data
sns.set_context("talk", font_scale=1.0)
seq_looms = results_df[results_df['Stim'].isin(['Loom 1', 'Loom 2', 'Loom 3'])].copy()

# 2. Generate the Base Grid Plot
g_min = sns.catplot(
    data=seq_looms,
    x='Stim',
    y='min',
    hue='Treatment Group',
    col='Region',
    col_wrap=3,
    kind='point',
    markers=['o', 's'],
    linestyles=['-', '--'],
    palette={'Vehicle': 'gray', 'Psilocybin': 'purple'},
    errorbar=('ci', 95),
    height=5, 
    aspect=1.2,
    sharey=False 
)

g_min.fig.suptitle('Negative Peak Adaptation: Vehicle vs Psilocybin', y=1.08, fontweight='bold')
g_min.set_axis_labels("Stimulus Sequence", "Amplitude (uV)")

# 3. The Statistical Annotation Loop
stimuli = ['Loom 1', 'Loom 2', 'Loom 3']

# zip() pairs each subplot axis (ax) with its specific region name
for ax, region in zip(g_min.axes.flat, g_min.col_names):
    
    # Invert the y-axis because 'min' LFP peaks are negative sinks
    ax.invert_yaxis()
    
    # Isolate the data for this specific region
    region_data = seq_looms[seq_looms['Region'] == region]
    
    # Get the mathematical limits of the y-axis to calculate our asterisk offset
    ymin, ymax = ax.get_ylim()
    y_range = abs(ymax - ymin)
    
    # Loop through each stimulus (0=Loom 1, 1=Loom 2, 2=Loom 3)
    for i, stim in enumerate(stimuli):
        # Extract purely the numerical 'min' values for both groups, dropping NaNs
        veh_data = region_data[(region_data['Stim'] == stim) & (region_data['Treatment Group'] == 'Vehicle')]['min'].dropna()
        psi_data = region_data[(region_data['Stim'] == stim) & (region_data['Treatment Group'] == 'Psilocybin')]['min'].dropna()
        
        # Only run stats if both groups have enough data points (n > 2)
        if len(veh_data) > 2 and len(psi_data) > 2:
            # Run Welch's t-test
            t_stat, p_val = ttest_ind(veh_data, psi_data, equal_var=False)
            
            # Determine significance level
            if p_val < 0.001:
                sig = '***'
            elif p_val < 0.01:
                sig = '**'
            elif p_val < 0.05:
                sig = '*'
            else:
                sig = '' # ns (not significant) - leave blank to avoid visual clutter
            
            if sig:
                # Calculate where to put the text.
                # Because the axis is inverted, the visually "highest" point is the mathematically lowest minimum.
                highest_point = min(veh_data.min(), psi_data.min())
                
                # Subtract 5% of the total y-axis range to bump the asterisk slightly UP visually
                text_y_pos = highest_point - (y_range * 0.05)
                
                # Draw the asterisk on the plot
                # x=i aligns it perfectly with Loom 1 (0), Loom 2 (1), etc.
                ax.text(x=i, y=text_y_pos, s=sig, ha='center', va='bottom', 
                        color='black', fontweight='bold', fontsize=18)

plt.tight_layout()
plt.show()

# Optional: Reset seaborn context
sns.set_context("notebook")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ---------------------------------------------------------
# PHASE 1: Data Extraction
# ---------------------------------------------------------
# 1. Extract the core metrics from the statsmodels 'result' object
# We build a new DataFrame directly from the model's internal arrays.
coef_df = pd.DataFrame({
    'Coefficient': result.params,
    'p_value': result.pvalues,
    'CI_Lower': result.conf_int()[0], # Index 0 is the 2.5% bound
    'CI_Upper': result.conf_int()[1]  # Index 1 is the 97.5% bound
})

# 2. Clean the Data
# The 'Intercept' and the random effects variance ('Group Var') operate on completely 
# different scales than the rest of the predictors. Plotting them squashes the variables we care about.
# We drop them from the dataframe using their index names.
coef_df = coef_df.drop(index=['Intercept', 'Group Var'], errors='ignore')

# 3. Clean up the messy statsmodels index names for the plot labels
# statsmodels creates ugly strings like "C(Q('Treatment Group'))[T.Psilocybin]"
# We use string replacement to make them readable.
clean_names = []
for name in coef_df.index:
    new_name = name.replace("C(Q('Treatment Group'))", "Treatment")
    new_name = new_name.replace("C(Region)", "Region")
    new_name = new_name.replace("C(Stim)", "Stim")
    new_name = new_name.replace("C(Experiment)", "Phase")
    new_name = new_name.replace("Q('latency min')", "Latency")
    new_name = new_name.replace("[T.", " [")
    new_name = new_name.replace("]", "]")
    clean_names.append(new_name)

coef_df.index = clean_names

# Sort values by Coefficient magnitude so the plot looks organized (staircase effect)
coef_df = coef_df.sort_values(by='Coefficient')

# ---------------------------------------------------------
# PHASE 2: Plotting the Forest Plot
# ---------------------------------------------------------
plt.figure(figsize=(10, 8))

# 1. Loop through the dataframe to plot each variable
# We use enumerate to get the y-coordinate (i) and the variable name (var_name)
for i, var_name in enumerate(coef_df.index):
    coef = coef_df.loc[var_name, 'Coefficient']
    lower = coef_df.loc[var_name, 'CI_Lower']
    upper = coef_df.loc[var_name, 'CI_Upper']
    p_val = coef_df.loc[var_name, 'p_value']
    
    # 2. Color Coding by Significance
    # If the p-value is < 0.05, we color it a bold red to draw the eye.
    # If it is not significant, we fade it to gray.
    if p_val < 0.05:
        color = 'firebrick'
        alpha = 1.0
    else:
        color = 'gray'
        alpha = 0.5
        
    # 3. Plot the Confidence Interval Line
    # plt.hlines draws a horizontal line from xmin (lower CI) to xmax (upper CI) at height y (i)
    plt.hlines(y=i, xmin=lower, xmax=upper, color=color, alpha=alpha, linewidth=3)
    
    # 4. Plot the Coefficient Dot
    # plt.plot draws the central point. 'o' specifies a circle marker.
    plt.plot(coef, i, 'o', markersize=8, color=color, alpha=alpha)
    
    # 5. Add Significance Asterisks
    # We place the text just slightly above the dot (i + 0.15)
    if p_val < 0.001: sig = '***'
    elif p_val < 0.01: sig = '**'
    elif p_val < 0.05: sig = '*'
    else: sig = ''
    
    if sig:
        plt.text(coef, i + 0.15, sig, ha='center', va='bottom', 
                 color=color, fontweight='bold', fontsize=14)

# ---------------------------------------------------------
# PHASE 3: Formatting the Axes and Null Line
# ---------------------------------------------------------
# Set the y-ticks to match our variables
plt.yticks(range(len(coef_df)), coef_df.index)

# Draw the critical Null Line at X=0
plt.axvline(x=0, color='black', linestyle='--', linewidth=1.5)

# Labels and Title
plt.xlabel('Effect Size on Negative Peak Amplitude (uV)\n<-- Stronger Sink (More Negative)  |  Weaker Sink (More Positive) -->', fontsize=12)
plt.title('Mixed-Effects Model Coefficients: Predictors of LFP Amplitude', fontweight='bold', fontsize=14, pad=20)

# Remove the top and right borders for a cleaner aesthetic (Tufte's data-ink ratio)
sns.despine()
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import statsmodels.formula.api as smf
import itertools

# =========================================================
# PHASE 1: Data Wrangling (Collapsing Micro to Macro)
# =========================================================

# 1. Create a function to group the regions based on their text prefixes
def group_macro_regions(region_name):
    # Ensure we are working with a string to prevent errors with NaNs
    if not isinstance(region_name, str):
        return region_name
    
    # Convert to uppercase for robust matching (catches 'SCzo', 'ScOp', 'scig')
    upper_name = region_name.upper()
    
    if upper_name.startswith('SC'):
        return 'SC'
    elif upper_name.startswith('PAG'):
        return 'PAG'
    else:
        # Keep any other regions (like MRN or Void) exactly as they are
        return region_name

# 2. Apply the function to create a brand new column, preserving the original
results_df['Macro_Region'] = results_df['Region'].apply(group_macro_regions)

# Verify it worked
print("Original Regions:", results_df['Region'].unique())
print("New Macro Regions:", results_df['Macro_Region'].unique())

# =========================================================
# PHASE 2: Re-fit the Model with the Explicit SC Baseline
# =========================================================

# Filter for the stimuli we care about and drop NAs for the required columns
target_stimuli = ['Loom 1', 'Loom 2', 'Loom 3']
model_df = results_df[results_df['Stim'].isin(target_stimuli)].copy()
cols_to_check = ['min', 'Treatment Group', 'Macro_Region', 'Stim', 'Experiment', 'latency min']
model_df = model_df.dropna(subset=cols_to_check)

# Define the formula
# CRITICAL CHANGE: We use Treatment(reference='SC') on the new Macro_Region column
# 1. Z-score the latency (mean of 0, standard deviation of 1)
model_df['latency_z'] = (model_df['latency min'] - model_df['latency min'].mean()) / model_df['latency min'].std()

# 2. Update the formula to use the new standardized variable
formula = (
    "Q('min') ~ C(Q('Treatment Group'), Treatment(reference='Vehicle')) * "
    "C(Stim, Treatment(reference='Loom 1')) + "
    "C(Q('Treatment Group')) * C(Experiment) + "
    "C(Macro_Region, Treatment(reference='SC')) + " 
    "latency_z"  # <--- Replaced Q('latency min')
)
# Fit the model
model = smf.mixedlm(formula, data=model_df, groups=model_df['Mouse'])
# Pass a list of methods so it tries 'lbfgs' first, then falls back to 'cg' if needed
# Increase maxiter to give it more time to resolve the curvature
result = model.fit(method=["lbfgs", "cg"], maxiter=5000)

# =========================================================
# PHASE 3: Generate the Estimated Marginal Means (EMMs)
# =========================================================

mean_latency = model_df['latency min'].mean()
treatments = ['Vehicle', 'Psilocybin']
stimuli = ['Loom 1', 'Loom 2', 'Loom 3']

grid = list(itertools.product(treatments, stimuli))
emm_df = pd.DataFrame(grid, columns=['Treatment Group', 'Stim'])

# Now we can safely hardcode 'SC' because it exists as a pure string in our new column!
emm_df['Macro_Region'] = 'SC'
emm_df['latency min'] = mean_latency
emm_df['Experiment'] = 'Extinction'

# Predict the pure voltages
emm_df['Predicted_Amplitude'] = result.predict(exog=emm_df)

print("\n--- Estimated Marginal Means (Extinction Phase, SC) ---")
print(emm_df[['Treatment Group', 'Stim', 'Predicted_Amplitude']])
```

```python
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.api as sm
import itertools

# 1. Clean Data & Z-Score Latency
target_stimuli = ['Loom 1', 'Loom 2', 'Loom 3']
model_df = results_df[results_df['Stim'].isin(target_stimuli)].copy()
cols_to_check = ['min', 'Treatment Group', 'Macro_Region', 'Stim', 'Experiment', 'latency min']
model_df = model_df.dropna(subset=cols_to_check)

# Z-score latency for numerical stability
model_df['latency_z'] = (model_df['latency min'] - model_df['latency min'].mean()) / model_df['latency min'].std()

# 2. The Formula
formula = (
    "Q('min') ~ C(Q('Treatment Group'), Treatment(reference='Vehicle')) * "
    "C(Stim, Treatment(reference='Loom 1')) + "
    "C(Q('Treatment Group')) * C(Experiment) + "
    "C(Macro_Region, Treatment(reference='SC')) + " 
    "latency_z"
)

# 3. Fit using GEE instead of MixedLM
# Note: groups is still required, but it is handled as a correlation structure
model = smf.gee(
    formula, 
    data=model_df, 
    groups=model_df['Mouse'], 
    cov_struct=sm.cov_struct.Exchangeable() # Assumes trials within a mouse are equally correlated
)

# Fit the model
result = model.fit()

# Print the output
print(result.summary())

# =========================================================
# PHASE 3: Generate the Estimated Marginal Means (EMMs)
# =========================================================
# GEE predicting works exactly the same way.
treatments = ['Vehicle', 'Psilocybin']
stimuli = ['Loom 1', 'Loom 2', 'Loom 3']

grid = list(itertools.product(treatments, stimuli))
emm_df = pd.DataFrame(grid, columns=['Treatment Group', 'Stim'])

emm_df['Macro_Region'] = 'SC'
# Because the model was trained on Z-scored latency, we must feed it a Z-score.
# A Z-score of 0 is exactly the mathematical mean.
emm_df['latency_z'] = 0.0 
emm_df['Experiment'] = 'Extinction'

emm_df['Predicted_Amplitude'] = result.predict(exog=emm_df)

print("\n--- Estimated Marginal Means (Extinction Phase, SC) ---")
print(emm_df[['Treatment Group', 'Stim', 'Predicted_Amplitude']])
```

```python
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.api as sm

# =========================================================
# PHASE 1: Split the Data (Extinction Only)
# =========================================================
target_stimuli = ['Loom 1', 'Loom 2', 'Loom 3']
cols_to_check = ['min', 'Treatment Group', 'Macro_Region', 'Stim', 'Experiment', 'latency min']

# Isolate Extinction data AND clean it in one step
ext_df = results_df[
    (results_df['Stim'].isin(target_stimuli)) & 
    (results_df['Experiment'] == 'Extinction')
].dropna(subset=cols_to_check).copy()

# Z-score latency for this specific subset
ext_df['latency_z'] = (ext_df['latency min'] - ext_df['latency min'].mean()) / ext_df['latency min'].std()

# =========================================================
# PHASE 2: The Streamlined Formula
# =========================================================
# Notice we removed 'Experiment' entirely because this model is ONLY Extinction data.
formula_ext = (
    "Q('min') ~ C(Q('Treatment Group'), Treatment(reference='Vehicle')) * "
    "C(Stim, Treatment(reference='Loom 1')) + "
    "C(Macro_Region, Treatment(reference='SC')) + " 
    "latency_z"
)

# =========================================================
# PHASE 3: Fit with Robust Independence
# =========================================================
# We switch to an 'Independence' working correlation. 
# GEE will STILL correct for the repeated measures (groups=Mouse) 
# using the robust sandwich estimator, but it won't force unstable covariance math.
model_ext = smf.gee(
    formula_ext, 
    data=ext_df, 
    groups=ext_df['Mouse'], 
    cov_struct=sm.cov_struct.Independence()
)

result_ext = model_ext.fit()
print("\n=== EXTINCTION PHASE MODEL ===")
print(result_ext.summary())
```

```python
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.api as sm

# =========================================================
# PHASE 1: Isolate and Prep Retention Data
# =========================================================
target_stimuli = ['Loom 1', 'Loom 2', 'Loom 3']
cols_to_check = ['min', 'Treatment Group', 'Macro_Region', 'Stim', 'Experiment', 'latency min']

# Isolate Retention data specifically
ret_df = results_df[
    (results_df['Stim'].isin(target_stimuli)) & 
    (results_df['Experiment'] == 'Retention')
].dropna(subset=cols_to_check).copy()

# Z-score the latency for this specific population
# Standardizing helps the GEE solver avoid the "invalid value encountered in sqrt" error
ret_df['latency_z'] = (ret_df['latency min'] - ret_df['latency min'].mean()) / ret_df['latency min'].std()

# =========================================================
# PHASE 2: The Streamlined Formula (No Experiment Variable)
# =========================================================
formula_ret = (
    "Q('min') ~ C(Q('Treatment Group'), Treatment(reference='Vehicle')) * "
    "C(Stim, Treatment(reference='Loom 1')) + "
    "C(Macro_Region, Treatment(reference='SC')) + " 
    "latency_z"
)

# =========================================================
# PHASE 3: Fit the GEE Model
# =========================================================
# We use the Independence correlation structure for maximum mathematical stability
model_ret = smf.gee(
    formula_ret, 
    data=ret_df, 
    groups=ret_df['Mouse'], 
    cov_struct=sm.cov_struct.Independence()
)

result_ret = model_ret.fit()

print("\n" + "="*30)
print("=== RETENTION PHASE MODEL (7 DAYS POST) ===")
print("="*30)
print(result_ret.summary())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------------------------------------
# PHASE 1: Data Extraction Function
# ---------------------------------------------------------
def extract_gee_results(model_result, phase_name):
    """Extracts coefficients and CIs from a GEE result object."""
    # 1. Get parameters and standard errors
    df = pd.DataFrame({
        'Coefficient': model_result.params,
        'Std_Error': model_result.bse,
        'p_value': model_result.pvalues
    })
    
    # 2. Calculate 95% Confidence Intervals (1.96 * SE)
    df['Lower_CI'] = df['Coefficient'] - (1.96 * df['Std_Error'])
    df['Upper_CI'] = df['Coefficient'] + (1.96 * df['Std_Error'])
    
    # 3. Label the phase
    df['Phase'] = phase_name
    return df

# ---------------------------------------------------------
# PHASE 2: Combine and Clean
# ---------------------------------------------------------
# Extract from both models
ext_coefs = extract_gee_results(result_ext, 'Extinction (1 Day)')
ret_coefs = extract_gee_results(result_ret, 'Retention (7 Days)')

# Merge them into one long-form DataFrame
all_coefs = pd.concat([ext_coefs, ret_coefs])

# Remove Intercept to keep the scale readable
all_coefs = all_coefs.drop(index=['Intercept'], errors='ignore')

# Clean up index names for the plot
all_coefs.index = [name.replace("C(Q('Treatment Group'), Treatment(reference='Vehicle'))", "Psilocybin")
                       .replace("C(Stim, Treatment(reference='Loom 1'))", "Stim")
                       .replace("C(Macro_Region, Treatment(reference='SC'))", "Region")
                       .replace("[T.", " [") for name in all_coefs.index]

# ---------------------------------------------------------
# PHASE 3: Plotting with FacetGrid
# ---------------------------------------------------------
sns.set_context("talk")
# height=6, aspect=1.2 ensures each plot is large enough to read
g = sns.FacetGrid(all_coefs.reset_index(), col="Phase", height=7, aspect=1.2, sharex=True)

def draw_forest_plot(data, **kwargs):
    """Custom function to draw the dots and lines on the grid."""
    # Sort so the variables appear in the same order on both plots
    data = data.sort_values('index')
    y_pos = range(len(data))
    
    # Draw the Null Line
    plt.axvline(0, color='black', linestyle='--', alpha=0.6)
    
    for i, row in enumerate(data.itertuples()):
        color = 'firebrick' if row.p_value < 0.05 else 'gray'
        alpha = 1.0 if row.p_value < 0.05 else 0.4
        
        # Horizontal error bar (Confidence Interval)
        plt.hlines(y=i, xmin=row.Lower_CI, xmax=row.Upper_CI, color=color, alpha=alpha, linewidth=3)
        # Central Dot (Coefficient)
        plt.plot(row.Coefficient, i, 'o', markersize=10, color=color, alpha=alpha)
        
        # Add asterisks for significance
        if row.p_value < 0.05:
            sig = '***' if row.p_value < 0.001 else '**' if row.p_value < 0.01 else '*'
            plt.text(row.Coefficient, i + 0.2, sig, ha='center', color=color, fontweight='bold')

    plt.yticks(y_pos, data['index'])

g.map_dataframe(draw_forest_plot)

# Final Polish
g.set_axis_labels("Effect Size (uV)", "")
g.fig.suptitle("Model Comparison: Extinction vs. Retention", y=1.05, fontweight='bold')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------------------------------------
# PHASE 1: Extract and Merge Model Parameters
# ---------------------------------------------------------
def get_model_params(result, phase):
    """Extracts parameters into a clean dataframe."""
    df = pd.DataFrame({
        'Variable': result.params.index,
        'Coef': result.params.values,
        'SE': result.bse.values,
        'P': result.pvalues.values,
        'Lower_CI': result.conf_int()[0].values,
        'Upper_CI': result.conf_int()[1].values,
        'Phase': phase
    })
    return df

# Extract from your fitted GEE models
ext_summary = get_model_params(result_ext, 'Extinction (Day 1)')
ret_summary = get_model_params(result_ret, 'Retention (Day 7)')

# Calculate Percent Change between Day 1 and Day 7
comparison = ext_summary.merge(ret_summary, on='Variable', suffixes=('_Ext', '_Ret'))

# Formula: ((New - Old) / |Old|) * 100
comparison['Percent_Change'] = ((comparison['Coef_Ret'] - comparison['Coef_Ext']) / abs(comparison['Coef_Ext'])) * 100

# Save this to CSV for your results section
comparison.to_csv('LFP_Longitudinal_Comparison.csv', index=False)
print("Summary table saved to 'LFP_Longitudinal_Comparison.csv'")

# ---------------------------------------------------------
# PHASE 2: Side-by-Side Forest Plot
# ---------------------------------------------------------
# Combine summaries for plotting
plot_df = pd.concat([ext_summary, ret_summary])

# Filter out Intercept and clean labels
plot_df = plot_df[plot_df['Variable'] != 'Intercept'].copy()
plot_df['Variable'] = plot_df['Variable'].str.replace("C(Q('Treatment Group'), Treatment(reference='Vehicle'))", "Psilocybin") \
                                         .str.replace("C(Stim, Treatment(reference='Loom 1'))", "Stim") \
                                         .str.replace("C(Macro_Region, Treatment(reference='SC'))", "Region") \
                                         .str.replace("[T.", " [")

sns.set_context("talk")
g = sns.FacetGrid(plot_df, col="Phase", height=8, aspect=1.2, sharex=True)

def forest_plot(data, **kwargs):
    # Sort for consistency
    data = data.sort_values('Variable')
    y_pos = np.arange(len(data))
    
    # Null line (No effect)
    plt.axvline(0, color='black', linestyle='--', alpha=0.5)
    
    for i, row in enumerate(data.itertuples()):
        # Significant variables are colored purple (Psilocybin theme)
        color = 'indigo' if row.P < 0.05 else 'darkgray'
        alpha = 1.0 if row.P < 0.05 else 0.4
        
        # Plot Confidence Intervals and Coefficients
        plt.hlines(y=i, xmin=row.Lower_CI, xmax=row.Upper_CI, color=color, alpha=alpha, linewidth=4)
        plt.plot(row.Coef, i, 'o', markersize=12, color=color, alpha=alpha)
        
        # Add labels
        if row.P < 0.05:
            sig = '***' if row.P < 0.001 else '**' if row.P < 0.01 else '*'
            plt.text(row.Coef, i + 0.2, sig, ha='center', fontweight='bold', color=color)

    plt.yticks(y_pos, data['Variable'])

g.map_dataframe(forest_plot)
g.set_axis_labels("Effect Size on LFP Amplitude (uV)", "")
g.fig.suptitle("Evolution of Neural Threat Processing: Day 1 vs Day 7", y=1.05, fontweight='bold', fontsize=20)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.api as sm
import itertools

# ---------------------------------------------------------
# PHASE 1: Data Preparation (Extinction & Retention)
# ---------------------------------------------------------
target_stimuli = ['Loom 1', 'Loom 2', 'Loom 3']

def prep_phase_df(df, phase):
    subset = df[(df['Stim'].isin(target_stimuli)) & (df['Experiment'] == phase)].copy()
    # Z-score latency for numerical stability
    subset['latency_z'] = (subset['latency min'] - subset['latency min'].mean()) / subset['latency min'].std()
    return subset.dropna(subset=['min', 'Treatment Group', 'Macro_Region', 'Stim', 'latency_z'])

ext_df_pag = prep_phase_df(results_df, 'Extinction')
ret_df_pag = prep_phase_df(results_df, 'Retention')

# ---------------------------------------------------------
# PHASE 2: Re-Fitting GEE with PAG Reference
# ---------------------------------------------------------
# We explicitly set Treatment(reference='PAG')
formula_pag = (
    "Q('min') ~ C(Q('Treatment Group'), Treatment(reference='Vehicle')) * "
    "C(Stim, Treatment(reference='Loom 1')) + "
    "C(Macro_Region, Treatment(reference='PAG')) + " 
    "latency_z"
)

# Fit Extinction
model_ext_pag = smf.gee(formula_pag, data=ext_df_pag, groups=ext_df_pag['Mouse'], 
                        cov_struct=sm.cov_struct.Independence())
result_ext_pag = model_ext_pag.fit()

# Fit Retention
model_ret_pag = smf.gee(formula_pag, data=ret_df_pag, groups=ret_df_pag['Mouse'], 
                        cov_struct=sm.cov_struct.Independence())
result_ret_pag = model_ret_pag.fit()

print("GEE Models re-fitted with PAG as the reference region.")
```

```python
# ---------------------------------------------------------
# PHASE 3: Estimated Marginal Means (PAG-specific)
# ---------------------------------------------------------
def get_pag_emms(result, phase_label):
    treatments = ['Vehicle', 'Psilocybin']
    stimuli = ['Loom 1', 'Loom 2', 'Loom 3']
    grid = list(itertools.product(treatments, stimuli))
    
    emm_df = pd.DataFrame(grid, columns=['Treatment Group', 'Stim'])
    emm_df['Macro_Region'] = 'PAG'  # We target the PAG
    emm_df['latency_z'] = 0.0       # Average latency
    emm_df['Predicted_Amplitude'] = result.predict(exog=emm_df)
    emm_df['Phase'] = phase_label
    return emm_df

emm_ext_pag = get_pag_emms(result_ext_pag, 'Extinction (Day 1)')
emm_ret_pag = get_pag_emms(result_ret_pag, 'Retention (Day 7)')
combined_emms = pd.concat([emm_ext_pag, emm_ret_pag])

# Plotting the PAG trajectories
sns.set_context("talk")
g = sns.FacetGrid(combined_emms, col="Phase", height=6, aspect=1.2, sharey=True)
g.map_dataframe(sns.pointplot, x="Stim", y="Predicted_Amplitude", hue="Treatment Group", 
                palette={'Vehicle': 'gray', 'Psilocybin': 'purple'}, markers=['o', 's'], linestyles=['-', '--'])

for ax in g.axes.flat:
    ax.invert_yaxis()

g.set_axis_labels("Stimulus", "Predicted PAG Amplitude (uV)")
g.fig.suptitle("PAG Fear Response: Vehicle vs Psilocybin", y=1.05, fontweight='bold')
plt.show()
```

```python
# ---------------------------------------------------------
# PHASE 4: Forest Plot (PAG Reference)
# ---------------------------------------------------------
def extract_params(result, phase):
    df = pd.DataFrame({'Coef': result.params, 'SE': result.bse, 'P': result.pvalues, 'Phase': phase})
    df['Lower'] = df['Coef'] - (1.96 * df['SE'])
    df['Upper'] = df['Coef'] + (1.96 * df['SE'])
    return df.drop('Intercept', errors='ignore')

plot_df = pd.concat([extract_params(result_ext_pag, 'Extinction'), 
                     extract_params(result_ret_pag, 'Retention')]).reset_index()

# Clean labels
plot_df['index'] = plot_df['index'].str.replace("C(Q('Treatment Group'), Treatment(reference='Vehicle'))", "Psilocybin") \
                                   .str.replace("C(Stim, Treatment(reference='Loom 1'))", "Stim") \
                                   .str.replace("C(Macro_Region, Treatment(reference='PAG'))", "Region") \
                                   .str.replace("[T.", " [")

plt.figure(figsize=(12, 8))
sns.pointplot(data=plot_df, x='Coef', y='index', hue='Phase', join=False, dodge=0.4, palette='viridis')

# Add error bars manually for the dodge
for i, row in enumerate(plot_df.itertuples()):
    # Calculate y-position with dodge
    y_idx = list(plot_df['index'].unique()).index(row.index)
    y_pos = y_idx - 0.2 if row.Phase == 'Extinction' else y_idx + 0.2
    plt.hlines(y=y_pos, xmin=row.Lower, xmax=row.Upper, color='black', alpha=0.5)

plt.axvline(0, color='black', linestyle='--')
plt.title("Effect Sizes relative to PAG Baseline", fontweight='bold')
plt.xlabel("Coefficient (uV difference from PAG)")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------------------------------------
# PHASE 1: Extract and Combine Data
# ---------------------------------------------------------
def get_clean_coefs(result, phase):
    df = pd.DataFrame({
        'Variable': result.params.index,
        'Coef': result.params.values,
        'P': result.pvalues.values,
        'Lower': result.conf_int()[0].values,
        'Upper': result.conf_int()[1].values,
        'Phase': phase
    })
    return df[df['Variable'] != 'Intercept']

# Extracting from your GEE models (now with PAG reference)
all_data = pd.concat([
    get_clean_coefs(result_ext_pag, 'Extinction (Day 1)'),
    get_clean_coefs(result_ret_pag, 'Retention (Day 7)')
])

# Professional Label Cleaning
all_data['Variable'] = all_data['Variable'].str.replace("C(Q('Treatment Group'), Treatment(reference='Vehicle'))", "Psilocybin") \
                                           .str.replace("C(Stim, Treatment(reference='Loom 1'))", "Stim") \
                                           .str.replace("C(Macro_Region, Treatment(reference='PAG'))", "Region") \
                                           .str.replace("[T.", " [")

# ---------------------------------------------------------
# PHASE 2: Custom Forest Plot Function
# ---------------------------------------------------------
sns.set_context("talk")
g = sns.FacetGrid(all_data, col="Phase", height=8, aspect=1.3, sharex=True)

def draw_beautiful_forest(data, **kwargs):
    # Sort variables so they are identical in both plots
    data = data.sort_values('Variable', ascending=False)
    y_pos = np.arange(len(data))
    
    # 1. Draw the Reference/Null Line
    plt.axvline(0, color='black', linestyle='--', linewidth=1.5, alpha=0.7)
    
    # 2. Iterate through each variable to plot
    for i, row in enumerate(data.itertuples()):
        # Significance coloring logic
        is_sig = row.P < 0.05
        color = 'firebrick' if is_sig else 'gray'
        alpha = 1.0 if is_sig else 0.4
        lw = 4 if is_sig else 2
        
        # Draw the Confidence Interval line
        plt.hlines(y=i, xmin=row.Lower, xmax=row.Upper, color=color, alpha=alpha, linewidth=lw)
        
        # Draw the Coefficient dot
        plt.plot(row.Coef, i, 'o', markersize=10, color=color, alpha=alpha, markeredgecolor='white')
        
        # Add Significance Stars
        if is_sig:
            sig_mark = '***' if row.P < 0.001 else '**' if row.P < 0.01 else '*'
            plt.text(row.Coef, i + 0.2, sig_mark, ha='center', va='bottom', 
                     color=color, fontweight='bold', fontsize=14)

    # Set Y-axis labels
    plt.yticks(y_pos, data['Variable'])

# ---------------------------------------------------------
# PHASE 3: Final Aesthetic Polish
# ---------------------------------------------------------
g.map_dataframe(draw_beautiful_forest)

# Labels with biological directionality
g.set_axis_labels("Difference from PAG Baseline (uV)\n<-- Stronger Sink | Weaker Sink -->", "")
g.fig.suptitle("PAG-Centered Model: Neural Effect of Psilocybin", y=1.05, fontweight='bold', fontsize=22)

# Remove the "spines" (box borders) for the Tufte-style clean look
sns.despine(left=True)
plt.tight_layout()
plt.show()
```
