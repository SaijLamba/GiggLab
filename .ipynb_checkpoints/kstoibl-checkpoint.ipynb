```python
import numpy as np
import pandas as pd
import scipy
import scipy.signal as signal
from scipy.stats import mode
from pathlib import Path
import cfad
import spikeinterface as si
import probe_to_xyz as pr
import pickle as pkl
import json


def process_and_load_alf(
    ks_dir, 
    raw_data_path, 
    output_dir, 
    triggers, 
    fs=30000.0, 
    n_channels=385,  # Updated default to match your error trace (385)
    dtype='int16', 
    gain_to_uvolts=0.195, 
    post_stim_buffer_s=0.5
):
    """
    Memory-safe conversion of Kilosort data to IBL ALF format.
    """
    
    ks_dir = Path(ks_dir)
    raw_data_path = Path(raw_data_path)
    output_dir = Path(output_dir)
    output_dir.mkdir(exist_ok=True, parents=True)
    
    print(f"--- Starting Processing ---")
    
    # =========================================================
    # 1. PROCESS TRIGGERS
    # =========================================================
    print(f"[1/5] Processing Triggers...")
    
    triggers = np.array(triggers)
    valid_intervals = []
    buffer_samples = int(post_stim_buffer_s * fs)

    if triggers.ndim == 1 and len(triggers) > 0:
        # Case A: 1D Sequence (Checkerboard)
        t_min = int(np.min(triggers))
        t_max = int(np.max(triggers) + buffer_samples)
        valid_intervals.append([t_min, t_max])
        print(f"  Input: 1D Array ({len(triggers)} events). Window: [{t_min}, {t_max}]")
        
    elif triggers.ndim == 2 and triggers.shape[1] == 2:
        # Case B: Explicit Windows
        valid_intervals = triggers.astype(int)
        print(f"  Input: 2D Array ({len(triggers)} windows).")
    else:
        raise ValueError(f"Invalid trigger shape: {triggers.shape}")

    all_intervals = np.vstack(valid_intervals)
    all_intervals = all_intervals[all_intervals[:, 0].argsort()]


    # =========================================================
    # 2. FILTER & SAVE SPIKES
    # =========================================================
    print(f"[2/5] Filtering Spikes...")
    
    def load_npy(name):
        p = ks_dir / name
        return np.load(p) if p.exists() else None

    st = load_npy('spike_times.npy')
    sc = load_npy('spike_clusters.npy')
    sa = load_npy('amplitudes.npy')
    sp = load_npy('spike_positions.npy')
    
    if st is None or sc is None:
        raise FileNotFoundError("spike_times.npy or spike_clusters.npy missing.")

    st = st.flatten()
    sc = sc.flatten()
    if sa is not None: sa = sa.flatten()

    keep_mask = np.zeros(len(st), dtype=bool)
    left_indices = np.searchsorted(st, all_intervals[:, 0], side='left')
    right_indices = np.searchsorted(st, all_intervals[:, 1], side='right')

    for l, r in zip(left_indices, right_indices):
        keep_mask[l:r] = True

    np.save(output_dir / 'spikes.times.npy', st[keep_mask] / fs)
    np.save(output_dir / 'spikes.clusters.npy', sc[keep_mask])
    if sa is not None: np.save(output_dir / 'spikes.amps.npy', sa[keep_mask])
    if sp is not None: np.save(output_dir / 'spikes.depths.npy', sp[keep_mask, 1])
    
    n_kept = np.sum(keep_mask)
    print(f"  Kept {n_kept} / {len(st)} spikes ({n_kept/len(st):.1%})")


    # =========================================================
    # 3. PROCESS CLUSTERS
    # =========================================================
    print(f"[3/5] Processing Clusters...")
    
    templates = load_npy('templates.npy')
    spike_temps = load_npy('spike_templates.npy')
    chan_pos = load_npy('channel_positions.npy')
    
    if (templates is not None) and (spike_temps is not None):
        unique_ids = np.unique(sc)
        c_depths, c_ptp = [], []
        
        for cid in unique_ids:
            mask = (sc == cid)
            if not np.any(mask):
                c_depths.append(np.nan); c_ptp.append(np.nan)
                continue
                
            try:
                m_res = mode(spike_temps[mask], axis=None, keepdims=False)
                modal_idx = m_res[0] if np.isscalar(m_res[0]) else m_res[0][0]
            except:
                modal_idx = spike_temps[mask][0]

            temp_data = templates[int(modal_idx), :, :]
            best_ch = np.argmax(np.ptp(temp_data, axis=0))
            
            c_depths.append(chan_pos[best_ch, 1] if chan_pos is not None else np.nan)
            wf = temp_data[:, best_ch]
            c_ptp.append(((np.argmax(wf) - np.argmin(wf)) / fs) * 1000.0)
            
        np.save(output_dir / 'clusters.depths.npy', np.array(c_depths))
        np.save(output_dir / 'clusters.peakToTrough.npy', np.array(c_ptp))
        
    metrics_df = pd.DataFrame()
    for f in ['cluster_KSLabel.tsv', 'cluster_ContamPct.tsv', 'cluster_Amplitude.tsv', 'cluster_group.tsv']:
        p = ks_dir / f
        if p.exists():
            df = pd.read_csv(p, sep='\t')
            if 'cluster_id' in df.columns: 
                metrics_df = pd.concat([metrics_df, df.set_index('cluster_id')], axis=1)
    
    if not metrics_df.empty:
        metrics_df.to_csv(output_dir / 'clusters.metrics.csv')


    # =========================================================
    # 4. RAW DATA QC (FIXED: Memory Safe Batching)
    # =========================================================
    print(f"[4/5] Running Raw Data QC (Memory Safe)...")
    
    if raw_data_path.exists():
        data_mem = np.memmap(raw_data_path, dtype=dtype, mode='r')
        full_data = data_mem.reshape(-1, n_channels)
        
        sos_ap = signal.butter(3, [300, min(10000, fs/2-1)], btype='bandpass', fs=fs, output='sos')
        sos_lf = signal.butter(3, [0.5, 300], btype='bandpass', fs=fs, output='sos')
        
        rms_ap_list, rms_lf_list, ts_list = [], [], []
        win_samples = int(fs * 1.0) # 1 second window
        
        for i_int, (start, end) in enumerate(all_intervals):
            duration = end - start
            if duration < win_samples: continue
                
            n_wins_total = duration // win_samples
            
            # --- MEMORY FIX: Loop strictly window-by-window ---
            # Using a loop here is efficient enough and guarantees low RAM usage.
            # Processing 1s of 385ch float32 data is ~45MB RAM. Safe.
            
            print(f"  Processing interval {i_int+1}/{len(all_intervals)} ({n_wins_total} windows)...")
            
            for w in range(n_wins_total):
                # Calculate offsets for this specific 1s window
                w_start = start + (w * win_samples)
                w_end = w_start + win_samples
                
                # Load ONLY 1 second of data
                chunk = full_data[w_start:w_end, :] # [30000, 385]
                
                # Convert to Volts (float32)
                chunk_volts = chunk.astype(np.float32) * gain_to_uvolts * 1e-6
                
                # Filter AP (axis 0 is time)
                filt_ap = signal.sosfiltfilt(sos_ap, chunk_volts, axis=0)
                val_ap = np.sqrt(np.mean(filt_ap**2, axis=0)) # Result [385,]
                
                # Filter LF
                filt_lf = signal.sosfiltfilt(sos_lf, chunk_volts, axis=0)
                val_lf = np.sqrt(np.mean(filt_lf**2, axis=0)) # Result [385,]
                
                rms_ap_list.append(val_ap)
                rms_lf_list.append(val_lf)
                ts_list.append(w_start / fs)
                
                if w % 50 == 0: 
                    # Simple progress indicator for long intervals
                    print(f"    Batch {w}/{n_wins_total}", end='\r')

        if rms_ap_list:
            np.save(output_dir / '_iblqc_ephysTimeRmsAP.rms.npy', np.vstack(rms_ap_list))
            np.save(output_dir / '_iblqc_ephysTimeRmsAP.timestamps.npy', np.array(ts_list))
            np.save(output_dir / '_iblqc_ephysTimeRmsLF.rms.npy', np.vstack(rms_lf_list))
            np.save(output_dir / '_iblqc_ephysTimeRmsLF.timestamps.npy', np.array(ts_list))
            
            # PSD Calculation (Random sampling)
            psd_accum = None; count = 0
            # Sample up to 20 random 4096-sample chunks from valid intervals
            valid_indices = np.random.choice(len(all_intervals), min(len(all_intervals), 20), replace=True)
            for idx in valid_indices:
                s, e = all_intervals[idx]
                if (e-s) < 2048: continue
                # Random start within the interval
                max_offset = (e - s) - 4096
                offset = np.random.randint(0, max_offset) if max_offset > 0 else 0
                
                chunk = full_data[s+offset : s+offset+4096, :].astype(np.float32) * gain_to_uvolts * 1e-6
                freqs, Pxx = signal.welch(chunk, fs=fs, nperseg=2048, axis=0)
                if psd_accum is None: psd_accum = np.zeros_like(Pxx)
                psd_accum += Pxx; count += 1
            
            if psd_accum is not None:
                np.save(output_dir / '_iblqc_ephysSpectralDensityLF.power.npy', psd_accum / count)
                np.save(output_dir / '_iblqc_ephysSpectralDensityLF.freqs.npy', freqs)
    else:
        print("  Warning: Raw data not found. Skipping QC.")


    # =========================================================
    # 5. BUILD TIDY DATAFRAME
    # =========================================================
    print(f"\n[5/5] Building Tidy DataFrame...")
    
    s_times = np.load(output_dir / 'spikes.times.npy')
    s_clust = np.load(output_dir / 'spikes.clusters.npy')
    s_amps  = np.load(output_dir / 'spikes.amps.npy') if (output_dir / 'spikes.amps.npy').exists() else None
    s_depth = np.load(output_dir / 'spikes.depths.npy') if (output_dir / 'spikes.depths.npy').exists() else None
    
    df = pd.DataFrame({'spike_time': s_times, 'cluster_id': s_clust})
    if s_amps is not None: df['spike_amp'] = s_amps
    if s_depth is not None: df['spike_depth'] = s_depth
    
    c_metrics = pd.read_csv(output_dir / 'clusters.metrics.csv') if (output_dir / 'clusters.metrics.csv').exists() else pd.DataFrame({'cluster_id': np.unique(s_clust)})
    if 'Unnamed: 0' in c_metrics.columns: c_metrics = c_metrics.rename(columns={'Unnamed: 0': 'cluster_id'})
    
    c_depths = np.load(output_dir / 'clusters.depths.npy')
    c_ptp = np.load(output_dir / 'clusters.peakToTrough.npy')
    
    if len(c_depths) == len(c_metrics): c_metrics['cluster_depth'] = c_depths
    if len(c_ptp) == len(c_metrics): c_metrics['cluster_duration_ms'] = c_ptp
    
    cols = [c for c in ['cluster_id', 'KSLabel', 'group', 'cluster_depth', 'cluster_duration_ms'] if c in c_metrics.columns]
    tidy_df = df.merge(c_metrics[cols], on='cluster_id', how='left')
    
    rename_map = {'KSLabel': 'quality', 'group': 'manual_group'}
    tidy_df = tidy_df.rename(columns=rename_map)
    
    print(f"--- Done. Returned DataFrame with {len(tidy_df)} rows. ---")
    return tidy_df
```
