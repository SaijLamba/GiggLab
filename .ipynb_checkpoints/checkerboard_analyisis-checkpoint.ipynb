```python
Psilo = vep_df[vep_df["Treatment Group"] == "Psilocybin"]["nVEP (µV)"]
Vehicle = vep_df[vep_df["Treatment Group"] == "Vehicle"]["nVEP (µV)"]
```

```python
print("Variance of vehicle is", np.var(Vehicle))
print("Variance of vehicle is", np.var(Psilo))
```

```python
stats.levene(Psilo, Vehicle)
```

```python
stats.shapiro(Vehicle)
```

```python
stats.shapiro(Psilo)
```

```python
stats.ttest_ind(Psilo, Vehicle, equal_var = True)
```
